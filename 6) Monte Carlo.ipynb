```python
# Example of Monte Carlo Integration in Python
import numpy as np
import matplotlib.pyplot as plt

repeatations=500
samplesize = 10000
a=0.0
b=1.0
factor = (b-a)/samplesize

def integrand(x):
  return x

results = []
for _ in range(repeatations):
  integral_result = 0.0
  for _ in range(samplesize):
    x = np.random.uniform(a, b)
    integral_result += integrand(x)
    MC=  factor*integral_result
  results.append(MC)

plt.title("Distribution")
plt.hist(results, bins=30, ec="black")
plt.xlabel("Estimated Values")
plt.show()
mean_result = np.mean(results)
std_dev = np.std(results)
min_result = np.min(results)
max_result = np.max(results)

print(f"Mean of the estimates: {mean_result:.5e}")
print(f"Standard deviation of the estimates: {std_dev:.5e}")
print(f"Minimum estimate: {min_result:.5e}")
print(f"Maximum estimate: {max_result:.5e}")
```
